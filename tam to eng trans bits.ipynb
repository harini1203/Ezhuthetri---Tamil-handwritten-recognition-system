```python

from translate import Translator

def translate_tamil_to_english(tamil_text):
    translator = Translator(to_lang="en", from_lang="ta")
    translation = translator.translate(tamil_text)
    return translation

if __name__ == "__main__":
    tamil_text = "வணக்கம்! இது ஒரு டெமோ ஆப்பிலிகேஷன் உதவி."
    translated_text = translate_tamil_to_english(tamil_text)
    print(f"Tamil: {tamil_text}")
    print(f"English: {translated_text}")
```
